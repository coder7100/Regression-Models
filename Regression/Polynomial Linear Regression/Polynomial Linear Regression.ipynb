# Polynomial Linear Regression

## Importing the Libraries

```python
%config IPcompleter.greedy= True
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Importing the Dataset

```python
data=pd.read_csv("Position_Salaries.csv")
#print(data)
X=data.iloc[:,1:-1]
y=data.iloc[:,-1]
#print(X)
#print(y)
```

## Training the Simple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
regressor1=LinearRegression()
regressor1.fit(X,y)
```

 ## Training the Polynomial Linear Regression Model

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=4)
X_poly=poly_reg.fit_transform(X)
regressor2=LinearRegression()
regressor2.fit(X_poly,y)
```

## Visualising the Simple Linear Regression model

```python
plt.scatter(X,y,color="red")
plt.plot(X,regressor1.predict(X),color="blue")
plt.title("Prediction Using Simple Linear Regression")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

## Visualising the Polynomial Regression Model

```python
plt.scatter(X,y,color="red")
plt.plot(X,regressor2.predict(X_poly),color="blue")
plt.title("Predicitng using Polynomial Regression Model")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```

## Predicting the Result using Simple Linear Regression

```python
regressor1.predict([[6.5]])
```

## Prediciting the Result Using Polynomial Linear Regression 

```python
regressor2.predict(poly_reg.fit_transform([[6.5]]))
```
